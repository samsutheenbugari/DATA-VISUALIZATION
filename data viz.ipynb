## univariant Discrete data=== Count -plot

```python
import seaborn as sns
```

```python
sns.countplot(data=df,x="cyl")
```

```python
plt.pie(x=df["cyl"].value_counts())
plt.show()
```

```python
plt.pie(x=df["cyl"].value_counts(),autopct="%1.0f%%")
plt.show()
```

```python
plt.pie(x=df["cyl"].value_counts(),autopct="%1.0f%%",labels=[8,4,6])
plt.show()
```

```python
plt.pie(x=df["cyl"].value_counts(),autopct="%1.0f%%",labels=['Blue',4,6],explode=[0.03,0.5,0.5])
plt.show()
```

```python
plt.pie(x=df["cyl"].value_counts(),autopct="%1.0f%%",labels=[8,4,6],explode=[0.03,0.5,0.5],shadow=True)
plt.show()
```
